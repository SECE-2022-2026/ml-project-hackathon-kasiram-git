```python
import pandas as pd
import numpy as np
```

```python
movies= pd.read_csv('movies.csv')
```

```python
movies.head()
```

```python
movies.info()
movies.columns
```

```python
movies['tags']= movies['genre'] + movies['overview']
movies.head()
```

```python
new_df= movies[['id', 'title', 'genre', 'overview', 'tags']]
new_df= new_df.drop(columns=['genre', 'overview'])
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=10000, stop_words='english')
cv
```

```python
vec= cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim= cosine_similarity(vec)
sim
```

```python
def recommend(movies):
  index= new_df[new_df['title']==movies].index[0]
  dist= sorted(list(enumerate(sim[index])), reverse=True, key=lambda vec:vec[1])
  for i in dist[0:5]:
    print(new_df.iloc[i[0]].title)
```

```python
movie_titles= movies['title']
movie_titles.head(20)
```

```python
recommend("The Good, the Bad and the Ugly")
```
